Conexão com API e alerta de status da requisição

```python
import requests
from plyer import notification

# URLs da API
api_url_banks = "https://brasilapi.com.br/api/banks/v1"
api_url_corretoras = "https://brasilapi.com.br/api/cvm/corretoras/v1"
api_url_pix = "https://brasilapi.com.br/api/pix/v1/participants"

# Chamada API e obtenção de resposta
response_bank = requests.get(api_url_banks)
response_corretoras = requests.get(api_url_corretoras)
response_pix = requests.get(api_url_pix)

# Validação e processamento dos dados da API Banks
if response_bank.status_code == 200:
    data_bank = response_bank.json()
    notification.notify(
        title='Requisição Banks Brasil API',
        message="Requisição concluída com sucesso. \nCódigo: " + str(response_bank.status_code),
        timeout=15
    )
    print("Código API Bank:", response_bank.status_code)
else:
    notification.notify(
        title='Erro na requisição Banks API',
        message="Houve um erro ao acessar a API. \nCódigo do erro: " + str(response_bank.status_code),
        timeout=10
    )
    print("Código API Bank:", response_bank.status_code)

# Validação e processamento dos dados da API Corretoras
if response_corretoras.status_code == 200:
    data_corretoras = response_corretoras.json()
    notification.notify(
        title='Requisição Corretoras Brasil API',
        message="Requisição concluída com sucesso. \nCódigo: " + str(response_corretoras.status_code),
        timeout=15
    )
    print("Código API Corretoras:", response_corretoras.status_code)
else:
    notification.notify(
        title='Erro na requisição Corretoras API',
        message="Houve um erro ao acessar a API. \nCódigo do erro: " + str(response_corretoras.status_code),
        timeout=10
    )
    print("Código API Corretoras:", response_corretoras.status_code)

# Validação e processamento dos dados da API PIX
if response_pix.status_code == 200:
    data_pix = response_pix.json()
    notification.notify(
        title='Requisição PIX Brasil API',
        message="Requisição concluída com sucesso. \nCódigo: " + str(response_pix.status_code),
        timeout=15
    )
    print("Código API PIX:", response_pix.status_code)
else:
    notification.notify(
        title='Erro na requisição PIX API',
        message="Houve um erro ao acessar a API. \nCódigo do erro: " + str(response_pix.status_code),
        timeout=10
    )
    print("Código API PIX:", response_pix.status_code)
```

Tabela 1 - Bancos no Brasil

```python
#Convertendo o JSon para dataFrame pandas
bancos_br = pd.DataFrame(data_bank)
bancos_br

#alterando os readers
bancos_br.rename(columns={'name':'nome_banco','code':'codigo','fullName':'nome_completo_do_banco'})
 
#Modificando os NaN para Sem Informação
bancos_br.fillna('Sem informação')
```

Tabela 2 - Corretoras no Brasil

```python
#Convertando o JSon para DataFrame Pandas
corretoras_br = pd.DataFrame(data_corretoras)
corretoras_br

#Modificando os NaN para Sem informação 
corretoras_br.fillna('Sem informação')
```
